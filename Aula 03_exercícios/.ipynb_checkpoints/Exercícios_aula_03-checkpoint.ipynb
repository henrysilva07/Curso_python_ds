```python
#Carregando o dataset
df = pd.read_csv(r'C:\Users\henry\repositorios_ds\Curso_python_ds\dataset\kc_house_data.csv')
```

### 1. Crie uma nova coluna chamada: “dormitory_type”
- Se o valor da coluna “bedrooms” for igual à 1 => ‘studio’
- Se o valor da coluna “bedrooms” for igual a 2 => ‘apartament’
- Se o valor da coluna “bedrooms” for maior que 2 => ‘house’

```python
df['dormitory_type'] = df['bedrooms'].apply(lambda x : "studio" if x == 1 else
                                           "apartament" if x == 2 else "house" )
```

### 2. Faça um gráfico de barras que represente a soma dos preços pelo número de quartos.

```python
# Primeiro vamos agrupar o preço por número de quartos
price_bedrooms = df[['price', 'bedrooms']].groupby('bedrooms').sum().reset_index()


sns.barplot( x = 'bedrooms' , y = 'price' , data = price_bedrooms );
plt.title("Soma do preço por tipo de média" , fontdict = {'fontsize': 20});



```

### 3. Faça um gráfico de linhas que represente a média dos preços pelo ano construção dos imóveis

```python
# Primeiro criando o grupo price x ano 
price_yr_built =  df[['price', 'yr_built']].groupby('yr_built').mean().reset_index()

sns.lineplot( x = 'yr_built' ,  y = 'price',  data = price_yr_built );
plt.title("Média dos preços dos imóveis por ano de construção" , fontdict = {'fontsize': 20});
```

### 4. Faça um gráfico de barras que represente a média dos preços pelo tipo dos dormitórios

```python
price_dormitory = df[['price', 'dormitory_type']].groupby("dormitory_type").mean().reset_index()

sns.barplot(x = 'dormitory_type' , y = 'price' , data  =  price_dormitory)

plt.title("Média de preços por tipo de dormitórios" , fontdict = {'fontsize': 20});

```

### 5. Faça um gráfico de linha que mostre a evolução da média dos preços pelo ano da reforma dos imóveis, a partir do ano de 1930.

```python
price_yr_renovated = df.loc[df['yr_renovated'] > 1930 , ['price' , 'yr_renovated']].groupby('yr_renovated').mean().reset_index()

sns.lineplot( x = 'yr_renovated' , y = 'price' , data = price_yr_renovated)

plt.title('Média de preço dos imóveis por ano de reforma a apartir de 1930', fontdict = {'fontsize': 20}); 
```

### 6. Faça um tabela que mostre a média dos preços por ano de construção e tipo de dormitórios dos imóveis

```python
df[['price' , 'yr_built' , 'dormitory_type']].groupby(['yr_built' , 'dormitory_type']).mean().reset_index()
```

### 7. Crie um Dashboard com os gráficos das questões 02, 03, 04 ( Dashboard: 1 Linha e 2 colunas )

```python
from matplotlib import gridspec

fig = plt.figure( figsize=(24, 12) ) # Criando a nossa figura (quadro) do dash

#Criando o número de grids(número de quadrados na figura) 
specs = gridspec.GridSpec( ncols=2, nrows=2, figure=fig )

ax1 = fig.add_subplot( specs[0, :] ) # primeira linha 
ax1.set_title('Gráfico 01')
ax2 = fig.add_subplot( specs[1, 0] ) # segunda linha e primeira coluna 
ax2.set_title('Gráfico 02')
ax3 = fig.add_subplot( specs[1, 1] ) # segunda linha e segunda coluna 
ax3.set_title('Gráfico 03')


#Questão 2
price_bedrooms = df[['price', 'bedrooms']].groupby('bedrooms').sum().reset_index()
sns.barplot( x = 'bedrooms' , y = 'price' , data = price_bedrooms , ax= ax2 );

#Questão 3
price_yr_built =  df[['price', 'yr_built']].groupby('yr_built').mean().reset_index()
sns.lineplot( x = 'yr_built' ,  y = 'price',  data = price_yr_built  , ax = ax1);

#Questão 4
price_dormitory = df[['price', 'dormitory_type']].groupby("dormitory_type").mean().reset_index()
sns.barplot(x = 'dormitory_type' , y = 'price' , data  =  price_dormitory , ax = ax3)
```
